```python
class Car:
    # Class variable
    num_wheels = 4

    def __init__(self, brand, model):
        # Instance variables
        self.brand = brand
        self.model = model

    def get_info(self):
        return f"This car is a {self.brand} {self.model} with {Car.num_wheels} wheels."


# Create two instances (objects) of the Car class
car1 = Car("Toyota", "Camry")
car2 = Car("Tesla", "Model S")

# Accessing instance variables and calling methods
print(car1.get_info())  # Output: "This car is a Toyota Camry with 4 wheels."
print(car2.get_info())  # Output: "This car is a Tesla Model S with 4 wheels."

# Accessing class variable
print(Car.num_wheels)  # Output: 4
```

**Q1.What is the primary goal of Object-Oriented Programming (OOP)?**

**Ans_01:**

The primary goal of Object-Oriented Programming (OOP) is to model and organize code in a way that closely mirrors real-world entities and their interactions. OOP is a programming paradigm that revolves around the concept of "objects," which are self-contained units containing both data (attributes) and behavior (methods).
The key objectives of OOP are:

* **Modularity and Reusability**: OOP promotes the creation of modular and reusable code components (objects). Once you define a class, you can create multiple instances (objects) of that class, making it easy to reuse code and reduce redundancy.

* **Encapsulation**: OOP emphasizes the concept of encapsulation, where the data and methods that operate on the data are bundled together within an object. This helps to hide the internal details of an object and expose only what is necessary for its interactions.

* **Abstraction:** OOP allows you to create abstract classes and methods that define a blueprint for objects without specifying their implementation. Abstraction enables programmers to focus on what an object does rather than how it does it, promoting a higher level of understanding and code readability.

* **Inheritance:** OOP supports the concept of inheritance, where one class (called the "subclass" or "derived class") can inherit the attributes and methods of another class (called the "superclass" or "base class"). Inheritance allows you to create hierarchical relationships between classes, facilitating code reuse and specialization.

* **Polymorphism:** OOP enables polymorphism, which allows different classes to be treated as objects of a common superclass. This enables more flexible and generalized coding, as the same code can be used for different classes, provided they share a common interface.

**Q4. What are attributes and methods in a class?**

**Ans_04:**

In object-oriented programming (OOP), a class is a blueprint or a template that defines the structure and behavior of objects. Classes have attributes and methods, which are essential components of a class.

**Attributes**

- Attributes are variables that belong to the class or its instances (objects).
- They represent the data or state associated with the class and its objects.
- Attributes define the characteristics or properties of the class or object.
- In Python, attributes are defined inside the class but outside any methods.
- They can be accessed using the object name or the class name (for class attributes).


**Methods:**
- Methods are functions defined within a class that define the behavior of the class or its instances.
- They define the actions or operations that the class or object can perform.
- Methods can access and modify the attributes of the class or object.
- In Python, the first parameter of a method is conventionally named self, which refers to the instance calling the method.
- Methods are accessed using the object name.

Attributes and methods work together to define the structure and behavior of objects in Python classes, making object-oriented programming a powerful way for modeling and organizing code in a more modular and reusable manner.

In Python, there are two types of variables that can be defined within a class: class variables and instance variables.

**Class Variables:**

- Class variables are shared among all instances (objects) of the class.
- They are defined inside the class but outside any methods.
- Class variables are accessed using the class name.

**Instance Variables:**
- Instance variables are specific to each instance (object) of the class.
- They are defined inside the class methods using the self keyword.
- Instance variables are accessed using the object (instance) name.

**Q6. What is the purpose of the self parameter in Python class methods?**

**Ans_06:**

**self parameter is a Reference to the Current Instance of the class.**


The self parameter in Python class methods is a special parameter that refers to the instance (object) of the class on which the method is called. It is a convention in Python to name the first parameter of a class method as self, though **you can use any valid variable name**.

The purpose of the self parameter is to allow class methods to access and manipulate the attributes and other methods of the object on which they are called. It acts as a reference to the current instance, enabling the method to work with the specific data associated with that instance.

When you call an instance method on an object, Python automatically passes the instance itself as the first argument to the method. This is why you need to include the self parameter in the method definition, even though you don't explicitly pass it when calling the method.


in **[Example 01](#Example_01)**, , the `get_info()` method accesses the instance attributes `self.brand` and `self.model` using the self parameter. When we call `car1.get_info()`, the self parameter implicitly refers to the `car1` object, allowing the method to access the specific attributes of that object.

Without the self parameter, the method would not know which instance's attributes to access, and it would not be able to differentiate between different objects of the class. The self parameter is a crucial element in enabling the concept of instance-specific data and behavior in Python classes.

**Q7.For a library management system, you have to design the "Book" class with OOP
principles in mind. The “Book” class will have following attributes:**

    a. title: Represents the title of the book.
    b. author: Represents the author(s) of the book.
    c. isbn: Represents the ISBN (International Standard Book Number) of the book.
    d. publication_year: Represents the year of publication of the book.
    e. available_copies: Represents the number of copies available for checkout.

**The class will also include the following methods:**

    a. check_out(self): Decrements the available copies by one if there are copies available for checkout.
    b. return_book(self): Increments the available copies by one when a book is returned.
    c. display_book_info(self):Displays the information about the book,including its attributes and the number of available copies
    
    
**Ans_07 - Please find below the code;**

```python
class Book:
    """
    WELCOME to Library Management System! Bellow are the Attributes;
    :param title: Represents the title of the book.
    :param author: Represents the author(s) of the book.
    :param isbn: Represents the ISBN (International Standard Book Number) of the book.
    :param publication_year: Represents the year of publication of the book.
    :param available_copies: Represents the number of copies available for checkout.
    """

    def __init__(self, title, author, isbn, publication_year, available_copies):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.publication_year = publication_year
        self.available_copies = available_copies

    def display_book_info(self):
        print("Book Information:")
        print(f"\t Title: {self.title}")
        print(f"\t Author(s): {self.author}")
        print(f"\t ISBN: {self.isbn}")
        print(f"\t Publication Year: {self.publication_year}")
        print(f"\t Available Copies: {self.available_copies} \n")

    def check_out(self):
        if self.available_copies > 0:
            self.available_copies -= 1
            print(f"Book '{self.title}' checked out. Now Available copies are : {self.available_copies}")
        else:
            print(f"SORRY!, NO available copies of '{self.title}' for checkout.")

    def return_book(self):
        self.available_copies += 1
        print(f"Book '{self.title}' returned. Now Available copies: {self.available_copies}")


def main():
    # Let's get the Book Class Information
    print(Book.__doc__)

    # Create Objects of Book Class and instantiate it.
    book_obj1 = Book("X-Men: God Loves, Man Kills", "Chris", "9788654552277", 1990, 5)
    book_obj2 = Book("Mike Tyson : Undisputed Truth", "Larry Sloman, Mike Tyson", "0964161484100", 1998, 10)
    book_obj3 = Book("V for Vendetta", "Alan Moore", "6901142585540", 2003, 6)

    # Lets Displays the information about the book,including its attributes and the number of available copies
    book_obj1.display_book_info()
    book_obj2.display_book_info()
    book_obj3.display_book_info()

    # lets use check_out() method
    book_obj1.check_out()
    book_obj1.check_out()
    book_obj1.check_out()
    book_obj1.check_out()
    book_obj1.check_out()
    book_obj1.check_out()
    
    # lets return_book() methods
    book_obj1.return_book()
    book_obj1.return_book()
    book_obj1.return_book()


if __name__ == "__main__":
    main()
```

**Q8. For a ticket booking system, you have to design the "Ticket" class with OOP principles in mind. The “Ticket” class should have the following attributes:**
    
    a. ticket_id: Represents the unique identifier for the ticket.
    b. event_name: Represents the name of the event.
    c. event_date: Represents the date of the event.
    d. venue: Represents the venue of the event.
    e. seat_number: Represents the seat number associated with the ticket.
    f. price: Represents the price of the ticket.
    g. is_reserved: Represents the reservation status of the ticket.
    
**The class also includes the following methods:**

    a. reserve_ticket(self): Marks the ticket as reserved if it is not already reserved.
    b. cancel_reservation(self): Cancels the reservation of the ticket if it is already reserved.
    c. display_ticket_info(self):Displays the information about the ticket, including its attributes and reservation status.
    
**Ans_08 - Please find below the code;**

```python
from uuid import uuid4


class Ticket:
    """
    WELCOME to Ticket Booking System! Bellow are the Attributes;
    :param ticket_id: Represents the unique identifier for the ticket.
    :param event_name: Represents the name of the event.
    :param event_date: Represents the date of the event.
    :param venue: Represents the venue of the event.
    :param seat_number: Represents the seat number associated with the ticket.
    :param price: Represents the price of the ticket.
    :param is_reserved: Represents the reservation status of the ticket.
    """

    # for a particular Event Below Attributes will be common hence declaring those as Class variable.
    EVENT_NAME = "A.R.Rahman Live Concert-New year Ev"
    EVENT_DATE = "31-Dec-2022"
    VENUE = "Royal Orchid-MUMBAI"

    def __init__(self, seat_number, price):
        self.ticket_id = str(0)  # Instance variable - Initializing with default value str(0)
        self.seat_number = seat_number
        self.price = price
        self.is_reserved = False  # Instance variable - Initializing with default value "False"

    def reserve_ticket(self):
        if not self.is_reserved:
            self.is_reserved = True
            self.ticket_id = str(uuid4())[:8]
            print(f"Seat Number {self.seat_number} Reserved Successfully,THANK YOU !!.")
        else:
            print(f"SORRY!! Seat {self.seat_number} is already reserved.")

    def cancel_reservation(self):
        if self.is_reserved:
            self.is_reserved = False
            print(f"Reservation for Seat {self.seat_number} has been Cancelled.")
        else:
            print(f"Seat {self.seat_number} is not reserved.")

    def display_ticket_info(self):
        print("Ticket Information:")
        print(f"\t Ticket ID         : {self.ticket_id}")
        print(f"\t Event Name        : {self.EVENT_NAME}")
        print(f"\t Event Date        : {self.EVENT_DATE}")
        print(f"\t Venue             : {self.VENUE}")
        print(f"\t Seat Number       : {self.seat_number}")
        print(f"\t Price             : ₹{self.price}")
        print(f"\t Reservation Status: {'Reserved' if self.is_reserved else 'Not Reserved/Cancelled'} \n")


# Let's get the Book Class Information
print(Ticket.__doc__)

# create a Object of Ticket Class
ticket1 = Ticket("Gold-A14", 250.00)

ticket1.reserve_ticket()
ticket1.display_ticket_info()
ticket1.reserve_ticket()

ticket1.cancel_reservation()
ticket1.display_ticket_info()

ticket1.reserve_ticket()
ticket1.display_ticket_info()

ticket2 = Ticket("Silver-J14", 180.00)
ticket2.reserve_ticket()
ticket2.display_ticket_info()
```

**Q9. You are creating a shopping cart for an e-commerce website. Using OOP to model
the "ShoppingCart" functionality the class should contain following attributes and
methods:**

    a. items: Represents the list of items in the shopping cart.
    
    The class also includes the following methods:
    a. add_item(self, item): Adds an item to the shopping cart by appending it to the list of items.
    b. remove_item(self, item): Removes an item from the shopping cart if it exists in the list.
    c. view_cart(self): Displays the items currently present in the shopping cart.
    d. clear_cart(self): Clears all items from the shopping cart by reassigning an empty list to the items attribute.
    
**Ans_09: Please Find below the code;**

```python
class ShoppingCart:

    def __init__(self):
        self.items = []

    def add_item(self, *cart_args, **cart_kwargs):
        """
        Designed this function to receive multiple arguments using *args and **kwargs concept
        :param cart_args:     Arbitrary Positional Arguments
        :param cart_kwargs:   Arbitrary Keyword Arguments
        :return: NA
        """
        for i in cart_args:
            self.items.append(i)
        for v in cart_kwargs.values():
            self.items.append(v)

    def remove_item(self, *cart_args, **cart_kwargs):
        # Check if any item is there in cart to remove or not.
        if len(self.items) > 0:
            removed_items = list(cart_args)
            removed_items.extend(cart_kwargs.values())
            # populate a list which will have valid items to be removed from cart
            removed_items = list(
                filter(lambda e: e in self.items, removed_items))
            for i in removed_items:
                self.items.remove(i)
            print(f"Following Items Removed from your Cart: {', '.join(removed_items)}")
        else:
            print(f"No Items in your cart to Remove. Cart is Empty!!")

    def view_cart(self):
        if len(self.items) > 0:
            print(f"Items in your Shopping Cart is:{self.items}")
        else:
            print(f"Your Shopping Cart is Empty!!{self.items}.")

    def clear_cart(self):
        self.items = []


def main():
    shopping_cart1 = ShoppingCart()
    shopping_cart1.view_cart()
    shopping_cart1.remove_item("MICROWAVE", x="SOFA")
    shopping_cart1.add_item("MOBILE")
    shopping_cart1.add_item("LAPTOP", "MICROWAVE")
    shopping_cart1.add_item(a="WEBCAM", b="TV")
    shopping_cart1.add_item("SOFA", "CHAIR", a="MIXER", b="KINDLE")
    shopping_cart1.view_cart()
    shopping_cart1.remove_item("MICROWAVE", "STOVE", x="SOFA")
    shopping_cart1.view_cart()
    shopping_cart1.clear_cart()
    shopping_cart1.view_cart()


if __name__ == "__main__":
    main()
```

    c. grade: Represents the grade or class of the student.
    d. student_id: Represents the unique identifier for the student.
    e. attendance: Represents the attendance record of the student.

**The class should also include the following methods:**

    a. update_attendance(self, date, status): Updates the attendance record of the student for a given date with the provided status (e.g., present or absent).
    b. get_attendance(self): Returns the attendance record of the student.
    c. get_average_attendance(self): Calculates and returns the average attendance percentage of the student based on their attendance record.
    
 **Ans_10: Please find below the code;**

```python
class SchoolERP:

    def __init__(self, name, age, grade, student_id):
        self.name = name
        self.age = age
        self.grade = grade
        self.student_id = student_id
        self.attendance = {}  # empty dictionary to save the student attendance

    def update_attendance(self, date, status):
        self.attendance[date] = status

    def get_attendance(self):
        print(f"#1.The Attendance Record of {self.name} is as follows:")
        for k, v in self.attendance.items():
            print(f"{k}:{v}")

    def get_average_attendance(self):
        present_count = list(self.attendance.values()).count("present")
        attendance_average = str(round(present_count / len(self.attendance) * 100, 2)) + "%"
        return f"The Average Attendance Percentage of {self.name} is: {attendance_average}"


def main():

    # Create a Object of SchoolERP Class for a student.
    john = SchoolERP("John", "14", "10th", 2825)

    # Lets upadate attendance of student
    john.update_attendance("2023-07-03", "present")
    john.update_attendance("2023-07-04", "present")
    john.update_attendance("2023-07-05", "present")
    john.update_attendance("2023-07-06", "present")
    john.update_attendance("2023-07-07", "absent")
    john.update_attendance("2023-07-10", "present")
    john.update_attendance("2023-07-11", "present")
    john.update_attendance("2023-07-12", "absent")
    john.update_attendance("2023-07-13", "absent")
    john.update_attendance("2023-07-14", "present")
    john.update_attendance("2023-07-15", "present")

    # Get the attendance Record of student
    john.get_attendance()

    # Get the average attendance percentage of student
    print("\n#2." + john.get_average_attendance())


if __name__ == "__main__":
    main()
```
