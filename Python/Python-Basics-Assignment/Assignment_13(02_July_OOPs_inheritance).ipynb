```python
# Import math Library
import math


class Shape:

    def __init__(self, colour, border_width):
        self.colour = colour
        self.border_width = border_width


class Rectangle(Shape):

    def __init__(self, colour, border_width, length, width):
        super().__init__(colour, border_width)
        self.length = length
        self.width = width

    def get_area(self):
        area = self.length * self.width
        print(f"The are of Rectangle having colour:{self.colour} = {area} ")


class Circle(Shape):

    def __init__(self, colour, border_width, radius):
        super().__init__(colour, border_width)
        self.radius = radius

    def get_area(self):
        area = 2 * math.pi * self.radius**2
        print(f"The are of Circle having colour:{self.colour} = {area} ")


def main():
    print("1.")
    shape1 = Shape("Red", 5.00)
    print(f"Attributes of Object shape1:", shape1.__dict__)

    print("2.")
    rectangle1 = Rectangle("BLACK", 6.00, 11.50, 14.45)
    print(f"Attributes of Object rectangle1:", rectangle1.__dict__)
    rectangle1.get_area()

    print("3.")
    circle1 = Circle("ORANGE", 7.00, 15.15)
    print(f"Attributes of Object circle1:", circle1.__dict__)
    circle1.get_area()


if __name__ == "__main__":
    main()
```

**Q1. Explain what inheritance is in object-oriented programming and why it is used.**

**Ans_01:**

Lets look at  **[Example_01](#Example_01)**  and  see below points;

Inheritance is an important aspect of the object-oriented paradigm. Inheritance provides code reusability to the program because we can use an existing class to create a new class instead of creating it from scratch.

In inheritance, the Derived class acquires the properties and can access all the data members and functions defined in the Base class. A Derived class can also provide its specific implementation to the functions of the Base class. 

Inheritance is a fundamental concept in object-oriented programming (OOP) that allows one class to inherit attributes and methods from another class. The class from which attributes and methods are inherited is called the superclass (or base class), and the class that inherits from the superclass is called the subclass (or derived class).

Inheritance is used to create a hierarchical relationship between classes, where the subclass is a specialized version of the superclass. This hierarchical structure allows you to model real-world relationships more effectively, as well as promote code reuse and organization. Here are the key aspects and benefits of inheritance:

**Code Reuse:** Inheritance enables you to reuse the code defined in the superclass in the subclass, avoiding duplication of code. By defining common attributes and methods in the superclass, you can have multiple subclasses that inherit and extend the behavior of the superclass. By using inheritance, we  can define common attributes and methods once in the `Shape` class and reuse them in the derived classes `(Rectangle and Circle)`. This avoids duplication of code and promotes a more organized codebase.

**Extensibility:** With inheritance, you can easily extend the functionality of existing classes by creating new subclasses. Subclasses can add new attributes and methods specific to their domain while inheriting the general characteristics from the superclass. If you want to add more **shapes** in the future, you can create new classes that inherit from Shape. You can add shape-specific attributes and methods in those new subclasses without modifying the existing code in the *Shape*, *Rectangle*, or *Circle classes*.

**Hierarchy and Organization:** Inheritance allows you to create a hierarchy of related classes, making the code more organized and easier to maintain. You can represent more complex relationships between classes using inheritance, leading to a more intuitive and modular code structure. Inheritance creates a clear hierarchy between classes, making it easier to understand the relationships between different `shapes`. The Shape class acts as a base class that defines common characteristics, and the derived classes extend or specialize the behavior.

**Polymorphism:** Inheritance facilitates polymorphism, which is the ability of objects to be treated as objects of a common superclass. This allows you to write generalized code that can handle different objects of related classes in a uniform manner.

The `get_area` method is an example of polymorphism. Both `Rectangle` and `Circle` classes have a method with the same name, but they implement it differently to calculate the area based on their specific attributes. The `main` function can use polymorphism to call `get_area` on different shape objects, treating them uniformly without knowing the specific shape.


**Specialization and Generalization**: Inheritance allows you to specialize classes, creating subclasses that are more specific or focused. you can create more general classes as superclasses to be inherited by more specialized subclasses.

Overall, inheritance allows you to create a more organized, flexible, and maintainable code structure, making it a powerful tool in object-oriented programming.

**Q2. Discuss the concept of single inheritance and multiple inheritance, highlighting their differences and advantages.**

**Ans_02:**

**Single Inheritance:**
Single inheritance is a form of inheritance in object-oriented programming where a class can inherit from only one base class (i.e., one parent class). In other words, a subclass can have only one direct parent class from which it inherits attributes and methods. This parent-child relationship creates a linear hierarchy. like below;

    class ParentClass:
        pass

    class ChildClass(ParentClass):
        pass

**Advantages of Single Inheritance**:

1. **Simplicity**: Single inheritance results in a straightforward and easy-to-understand class hierarchy. There's only one direct parent class, so there's no ambiguity about where attributes and methods come from.

2. **Code Reusability**: Single inheritance still allows for code reuse, as the subclass inherits all the attributes and methods from its single base class. This promotes cleaner and more modular code.

3. **Avoiding Diamond Problem**: In single inheritance, there's no possibility of the "diamond problem," which can occur in multiple inheritance (discussed later). The diamond problem arises when a class inherits from multiple classes that have a common ancestor, potentially leading to ambiguity in method resolution.

**Multiple Inheritance:**
Multiple inheritance is a form of inheritance in object-oriented programming where a class can inherit from multiple base classes (i.e., more than one parent class). The subclass can have multiple direct parent classes, and it inherits attributes and methods from all of them.

    class ParentClass1:
        pass

    class ParentClass2:
        pass

    class ChildClass(ParentClass1, ParentClass2):
        pass

**Advantages of Multiple Inheritance:**

1. **Code Reusability and Modularity**: Multiple inheritance allows a class to inherit attributes and methods from multiple base classes. This results in better code reusability and modularity, as a class can combine features from different classes to achieve specific functionality.

2. **Interface Composition:** By inheriting from multiple classes, a subclass can implement different interfaces, each defining a specific set of attributes and methods. This allows for better segregation of functionality and promotes the composition of classes to create more complex behaviors.

3. **Expressive and Flexible:** Multiple inheritance allows for more expressive and flexible class relationships. It can represent complex real-world relationships where a class shares features from multiple sources.

**Differences:**
1. **Number of Base Classes:** The primary difference between single and multiple inheritance is the number of base classes a subclass can have. Single inheritance allows only one base class, while multiple inheritance allows multiple base classes.

2. **Class Hierarchy:** In single inheritance, the class hierarchy is linear, and there's a clear parent-child relationship between classes. In multiple inheritance, the class hierarchy becomes more complex, as classes can have multiple parent classes.

3. **Method Resolution Order:** In multiple inheritance, the order in which base classes are listed in the subclass matters. Python follows the C3 linearization algorithm (C3 superclass linearization) to determine the method resolution order (MRO) when dealing with multiple inheritance.

**Summary:**
While multiple inheritance provides more flexibility, it can also introduce complexity and potential pitfalls like the diamond problem. In contrast, single inheritance offers simplicity and avoids some of the complications associated with multiple inheritance. The choice between single and multiple inheritance depends on the specific design requirements of the application. In many cases, single inheritance is sufficient, but multiple inheritance can be valuable in more complex scenarios where class composition and combination of features from multiple sources are required.

**Q4. What is the significance of the "protected" access modifier in inheritance? How does it differ from "private" and "public" modifiers?**

**Ans_04:**



in Python, the **"protected"** access modifier is a convention rather than a strict enforcement of access control. It is indicated by a single underscore prefix (e.g., `_protected_attribute`) for attributes and methods in a class that should be treated as protected. This means that they are intended for internal use within the class and its subclasses but should not be accessed or modified directly from outside the class hierarchy.

A single leading underscore '_' in front of a variable, a function, or a method name means that these objects are used internally (i.e. protacted access Modifier). It is more of a syntax hint to the programmer and is not a Rule by the Python interpreter which means these objects can still be accessed by couple of ways BUT should be avoided.

However, Python does not enforce strict access control, and all attributes and methods are technically accessible from outside the class. The use of the single underscore prefix is a way to indicate the intended access level and suggest that these members should be considered as implementation details.

**Differences between "protected," "private," and "public" access modifiers:**

1. **Public Access Modifier:**
- In Python, attributes and methods without any underscore prefix are considered "public."
- Public members can be accessed and modified from anywhere, including outside the class and its subclasses.
2. **Protected Access Modifier:**
- In Python, attributes and methods with a `single underscore(_)` prefix (e.g., _protected_attribute) are considered "protected."
- Protected members are meant to be used within the class and its subclasses but can technically be accessed from outside the class as well.
- The single underscore prefix serves as a convention to indicate that the attribute or method is intended for internal use and should not be accessed directly by external code.
3.  **Private Access Modifier:**
- In Python, attributes and methods with a `double underscore(__)` prefix (e.g., __private_attribute) are considered "private."
- Private members are intended to be used only within the class itself and cannot be accessed directly from outside the class, including its subclasses.
- Python enforces name mangling to make private attributes more difficult to access accidentally from external code.


**Note:**  The access control in Python is not as strict as in some other programming languages. The single underscore for protected members and double underscore for private members are just conventions to signal the intended access level. Developers are responsible for following these conventions, as Python will not prevent direct access to protected or private members.
It's essential to use the appropriate access modifiers and follow the conventions to ensure proper encapsulation and maintainability of the code. The access level should be chosen based on the desired level of encapsulation and whether the attributes or methods need to be accessed outside the class hierarchy.




**Q5. What is the purpose of the "super" keyword in inheritance? Provide an example.**

**Ans_05:**

Refer **[Example_01](#Example_01)**  and see below points;

The **"super"** keyword in inheritance is used to call a method of the parent class (base class) within a method of the child class (derived class). It allows the child class to access and utilize the behaviour defined in the parent class, promoting code reusability and maintaining the hierarchical relationship between classes.

In the provided example, we can use the **"super"** keyword to call the constructor (`__init__` _ method) of the base class `Shape` from the derived classes `Rectangle` and `Circle`. This ensures that the common attributes (`colour` and `border_width`) defined in the base class are initialized correctly when creating objects of the derived classes.

we use the "super" keyword within the __init__ method of both Rectangle and Circle classes. When super().__init__(colour, border_width) is called, it refers to the constructor of the base class Shape. This way, the colour and border_width attributes are initialized appropriately for each object created from the derived classes.

By using the "super" keyword, we ensure that any changes or updates made to the `Shape` class's constructor will automatically be reflected in the derived classes without having to modify their code. It enhances code maintainability and promotes a clear and organized inheritance hierarchy.

**NOTE – in real time projects we don’t have to bother name of Parent class  in such case `super()` is very useful**



**Q6. Create a base class called "Vehicle" with attributes like "make", "model", and "year".
Then, create a derived class called "Car" that inherits from "Vehicle" and adds an
attribute called "fuel_type". Implement appropriate methods in both classes.**

```python
class Vehicle:

    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year


class Car(Vehicle):

    def __init__(self, make, model, year, fuel_type):
        super().__init__(make, model, year)
        self.fuel_type = fuel_type


def main():
    car1 = Car("BMW", "X3", "2020", "PETROL")
    car2 = Car("Tesla", "S", "2021", "EV")
    print(f"Attributes of Object car1:", car1.__dict__)
    print(f"Attributes of Object car2:", car2.__dict__)


if __name__ == "__main__":
    main()
```

**Q10. Create a base class called "BankAccount" with attributes like "account_number" and
"balance." Derive two classes, "SavingsAccount" and "CheckingAccount," from
"BankAccount." Add specific methods like "calculate_interest" for the
"SavingsAccount" class and "deduct_fees" for the "CheckingAccount" class.**

**Ans_10:**

Please check below code ;

**NOTE-**

 In the Below example, there is no need to use **super()** because the constructor `__init__` in the derived classes does not introduce any additional attributes beyond those already defined in the base class (BankAccount). Therefore, there's **no need to explicitly call the base class's constructor.**

When the `__init__` method is not defined in a subclass, Python automatically looks for it in the base class and calls it if found. This is known as implicit inheritance of the constructor. Since the `__init__` method is present in the BankAccount class, creating instances of the derived classes (SavingsAccount and CheckingAccount) will automatically invoke the constructor of the base class.

However, if you had additional attributes to initialize in the derived classes or needed to perform some custom initialization along with the base class initialization, you could use **super()** to call the base class's constructor explicitly. 


```python
class BankAccount:

    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance


class SavingsAccount(BankAccount):

    def calculate_interest(self, rate):
        interest = self.balance * rate / 100
        self.balance += interest
        print(f"Interest calculated: {interest}")


class CheckingAccount(BankAccount):

    def deduct_fees(self, fee):
        if self.balance >= fee:
            self.balance -= fee
            print(f"Fees deducted: {fee}")
        else:
            print("Insufficient balance for fees deduction!")


def main():
    print("1.")
    bankaccount1 = BankAccount(123456, 8500.00)
    print(f"Attributes of Object bankaccount1:", bankaccount1.__dict__)
    print("2.")
    savingsaccount1 = SavingsAccount(123456, 8500.00)
    print(f"Attributes of Object savingsaccount1:", savingsaccount1.__dict__)
    savingsaccount1.calculate_interest(2.5)
    print(f"Attributes of Object savingsaccount1:", savingsaccount1.__dict__)
    print("3.")
    checkingaccount1 = CheckingAccount(123456, 8500.00)
    print(f"Attributes of Object checkingaccount1:", checkingaccount1.__dict__)
    checkingaccount1.deduct_fees(100)
    print(f"Attributes of Object checkingaccount1:", checkingaccount1.__dict__)


if __name__ == "__main__":
    main()
```
